# Generate Random Data for Training and Test
To create our random data, I created a function that takes in the input state and outputs one of the specified states:
 * state 1  =>  `0011`
 * state 2  =>  `0101`
 * state 3  =>  `1010`
 * state 4  =>  `1100`

```python
def get_state(x):
    if x == 1:
        return '0011'
    elif x == 2:
        return '0101'
    elif x == 3:
        return '1010'
    elif x == 4:
        return '1100'
```

Generate 360 random input output states. As described in the problem, there is the question asked `What would happen if you provided a different state?`. To answer this, I added another feature that is `1` if the input state is `1`, `2`, `3`, or `4` because that was what was specified in the problem. If it is not one for those, that feaure is 0.

```python
X0 = [np.random.randint(1, 5) for i in range(360)]
X1 = [1 if x in [1, 2, 3, 4] else 0 for x in X0]
y = [get_state(x) for x in X0]
df = pd.DataFrame({
    'X0': X0,
    'X1': X1,
    'y': y
})
df.head()
```

# Data Preprocessing
Date preprocessing is one of the most important things in quantum machine learning because we need to convert/encode our data to be about to be inputted into our quanutum circuit. I created a function to preprocess input states when testing on different values to ease the process of testing.

First I normalize the features with `scikit-learn` `standardscaler` which normalizes the values between 0 and 1. The `scikit-learn` `minmaxscaler` does what it sounds like. This is done because the quamtum circuit can take values between `-1` and `1`. I added other inputs for the preprocessing so then the scaling would perform better when testing on input values not in the dataset of `1 to 4`.

```python
X_train, X_test, y_train, y_test = train_test_split(df[['X0', 'X1']], df['y'], test_size=0.2)

outside_data = [[x, 0] for x in range(-7, 1)] + [[x, 0] for x in range(5, 11)]

# Normalize
std_scale = StandardScaler().fit(np.append(X_train, outside_data, axis=0))
X_train = std_scale.transform(X_train)
X_test = std_scale.transform(X_test)

# Scale
X = np.append(X_train, X_test, axis=0)
minmax_scale = MinMaxScaler((-1, 1), clip=True).fit(np.append(X, outside_data, axis=0))
X_train = minmax_scale.transform(X_train)
X_test = minmax_scale.transform(X_test)

# Select
train_size = 100
X_train = X_train[:train_size]
y_train = y_train[:train_size].to_numpy()

test_size = 20
X_test = X_test[:test_size]
y_test = y_test[:test_size].to_numpy()

def preprocess_input(x):
    x = [[x, 1 if x in [1, 2, 3, 4] else 0]]
    x = std_scale.transform(x)
    x = minmax_scale.transform(x)
    return x
```

```python
print(X_train[0], y_train[0])
print(X_test[0], y_test[0])
```

# Data Encoding

```python
zz_map = ZZFeatureMap(feature_dimension=2, reps=2, entanglement='linear', insert_barriers=True)
zz_kernel = QuantumKernel(feature_map=zz_map, quantum_instance=Aer.get_backend('statevector_simulator'))
```

```python
print(X_train[0])
print(X_train[1])
```

```python
zz_circuit = zz_kernel.construct_circuit(X_train[0], X_train[1])
zz_circuit.decompose().decompose().draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(zz_circuit, backend, shots=8192,
              seed_simulator=1024, seed_transpiler=1024)
counts = job.result().get_counts(zz_circuit)
```

```python
plot_histogram(counts)
```

```python
matrix_train = zz_kernel.evaluate(x_vec=X_train)
matrix_test = zz_kernel.evaluate(x_vec=X_test, y_vec=X_train)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(np.asmatrix(matrix_train),
              interpolation='nearest', origin='upper', cmap='Blues')
axs[0].set_title('training kernel matrix')
axs[1].imshow(np.asmatrix(matrix_test),
              interpolation='nearest', origin='upper', cmap='Reds')
axs[1].set_title('testing kernel matrix')
plt.show()
```

# Quantum Support Vector Classification
To predict the value for each of the possible inputs, I used scikit-learn's Support Vector Classification. It is fitted using the `zz_kernel` specified above. The SVC is done classically unlike the other steps. The quantum kernel support vector classification algorithm was introduced in [Supervised learning with quantum enhanced feature spaces - Havlicek et al. Nature 567, 209-212 (2019)](https://arxiv.org/abs/1804.11326).

```python
zzcb_svc = SVC(kernel=zz_kernel.evaluate)
zzcb_svc.fit(X_train, y_train)
zzcb_score = zzcb_svc.score(X_test, y_test)

print(f'Kernel Classification Test Score: {zzcb_score}')
```

```python
print(X_test[0], y_test[0])
```

```python
zzcb_svc.predict([X_test[0]])
```

To test the classification on other inputs `(not 1, 2, 3, or 4)`, the preprocessing step sets the second feature to a `0` as mentioned earlier. This then gets scaled to a `-1` after preprocessing.

```python
preprocess_input(3), zzcb_svc.predict(preprocess_input(3))
```

```python
print(f'Input \t \t Preprocessed \t \t Prediction')
for x in range(-9, 20):
    prediction = zzcb_svc.predict(preprocess_input(x))
    print(f'{x} \t {preprocess_input(x)}; \t {prediction}')
```

Ideally, we would want different outputs outside this range, but increasing the range in preprocessing, decreases the overall accuracy of the `svc` model.
